```python
import numpy as np
import scipy.optimize as sopt
def f(x):
 return (x[0]-4)**4 + (x[1]-3)**2 + 4*(x[2]+5)**4
def del_f(x):
 return np.array([4*(x[0]-4)**3, 2*(x[1]-3), 16*(x[2]+5)**3])
def pi(alpha):
  return f(x - alpha*del_f(x))
```

```python
X,epsilon= [np.array([4, 2,-1])],10**(-6)
while True:
    x = X[-1]
    alpha_opt = sopt.golden(pi)
    next_x = x - alpha_opt * del_f(x)
    X.append(next_x)
    print("updated x",next_x)
    if np.linalg.norm(next_x-x)/np.linalg.norm(x) <epsilon:
        break
```

```python
import numpy as np
import scipy.optimize as sopt
def f(x):
 return x[0]+ x[1]/2 + (x[0]**2)/2+x[1]**2+3
def del_f(x):
 return np.array([1+x[0],0.5+2*x[1]])
def pi(alpha):
  return f(x - alpha*del_f(x))
```

```python
epsilon = 10**(-6)
X= [np.array([0,0])]
for i in range(2):
    x = X[-1]
    alpha_opt = sopt.golden(pi)
    print("alpha",alpha_opt)
    next_x = x - alpha_opt * del_f(x)
    X.append(next_x)
    #print("alpha_opt",alpha_opt)
    print("updated x",next_x)
    #print("next_x-x",next_x-x)
    #print(np.linalg.norm(next_x-x),"norm임")
    #print("")
    if np.linalg.norm(next_x-x)/np.linalg.norm(x) <epsilon:
        break
```
