```python
def find_next(arr, num, con): #연속되는 갯수를 반환하는 함수
        for i in range (len(arr)):
                if arr[i]==num+1:
                        num += 1
                        con += 1
                        return find_next(arr, num, con) #재귀호출
        return con; #없을경우 현재 최대 연속수인 con을 반환
```
