```python
from sklearn.ensemble import ExtraTreesClassifier

data= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/인공지능의이해/train.csv')
data
```

```python
X = data.iloc[:,0:6]
X 
```

```python
y= data.iloc[:,-1] # 마지막 컬럼
y
```

```python
model = ExtraTreesClassifier()
model.fit(X, y)

print(model.feature_importances_)

feat_importances = pd.Series(model.feature_importances_,index=X.columns)
feat_importances.nlargest(5).plot(kind = 'barh')
plt.show()
```

```python
#quiz 
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
from sklearn.ensemble import ExtraTreesClassifier

data= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/인공지능의이해/diabetes.csv')
```

```python
data = data[data.columns[data.isnull().mean()<0.3]]
```

```python
data = data.fillna(data.mean())
data
```

```python
X = data.iloc[:,0:7]
X

y= data.iloc[:,-1] # 마지막 컬럼
y
model = ExtraTreesClassifier()
model.fit(X, y)

print(model.feature_importances_)

feat_importances = pd.Series(model.feature_importances_,index=X.columns)
feat_importances.nlargest(7).plot(kind = 'barh')
plt.show()
```
